```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(
    "car_evaluation.csv",
    header=None,
    names=['buying','maint','doors','persons','lug_boot','safety','class']
)
df.shape
```

## In this dataset header is not written so If you loaded the car dataset without fixing headers, pandas may behave oddly.

```python
df = df.iloc[:,[0,1,4,5,6]]
df.sample(5)
```
